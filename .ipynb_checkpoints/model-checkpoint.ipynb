```python
import numpy as np
import pandas as pd
from datetime import datetime
from sklearn.cluster import KMeans, DBSCAN, AgglomerativeClustering, SpectralClustering, OPTICS
from sklearn.mixture import GaussianMixture
from sklearn.preprocessing import PowerTransformer, RobustScaler, MinMaxScaler
from sklearn.decomposition import PCA
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
df = pd.read_csv('/home/athel/Desktop/Litmus7/order/MergedData.csv')
df['Date'] = pd.to_datetime(df['Date'])
df = df.sort_values(['Customer ID', 'Date'])
df['Order Gap'] = df.groupby('Customer ID')['Date'].diff().dt.days
df['Order Gap'] = df['Order Gap'].fillna(0)
df.head()
```

```python
df['Sales'] = df['Quantity'] * df['Price per Unit']
```

```python
today = pd.to_datetime(datetime.now().date())
df['Recency'] = (today - df['Date']).dt.days
```

```python
customer_df = df.groupby('Customer ID').agg(
    total_spend = ('Sales', 'sum'),
    purchase_frequency = ('Order ID', 'nunique'),
    avg_basket_size = ('Quantity', 'mean'),
    cat_diversity = ('Category','nunique'),
    recency = ('Date', lambda x: (pd.Timestamp.today() - x.max()).days),
    gap = ('Order Gap', lambda x: x.sum() / (np.count_nonzero(x) + 1) if np.count_nonzero(x) >= 0 else None),
    age = ('Customer Age', 'first')
).reset_index()

print(df[df['Customer ID'] == 'CUST001']['Order Gap'])
customer_df
```

```python
features = ['total_spend','purchase_frequency','avg_basket_size','cat_diversity','age','recency','gap']
X = customer_df[features].values
X
```

```python
pt = PowerTransformer(method='yeo-johnson', standardize=True)
pca = PCA(n_components=2)
# Fit and transform in sequence
X_pt = pt.fit_transform(X)
X_pca = pca.fit_transform(X_pt)
```

```python
plt.figure(figsize = (15,6))

plt.subplot(1,3,1)
sns.boxplot(x = customer_df['total_spend'])
plt.title('Spend Boxplot')
plt.xlabel('Total Spend')

plt.subplot(1,3,2)
plt.boxplot(x = customer_df['purchase_frequency'])
plt.title('Frequency Boxplot')
plt.xlabel('Frequency')


plt.subplot(1,3,3)
plt.boxplot(x = customer_df['recency'])
plt.title ('Recency Boxplot')
plt.xlabel('Recency')

plt.tight_layout()
plt.show()

customer_df.nsmallest(1, 'recency')
```

# KMeans

```python
kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['KMeans Cluster'] = kmeans.fit_predict(X_pca)

# Plot
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=customer_df['KMeans Cluster'], palette='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans')
plt.show()

```

```python
loadings = pca.components_

feature_names = ['total_spend', 'purchase_frequency', 'avg_basket_size','cat_diversity','age','recency','gap']
loadings_df = pd.DataFrame(loadings.T, columns = ['PC1', 'PC2'], index = feature_names)

print(loadings_df)
print(pca.explained_variance_ratio_)
```

PC1 holds more weightage from total_spend, purchase_frequency and cat_diversity, while PC2 holds more weightage from avg_basket_size predominantly. Recency, however, has very negligible relevance according to this model.

# DBScan

```python
dbscan = DBSCAN(eps = 0.5, min_samples = 5)
customer_df['DBScan Cluster'] = dbscan.fit_predict(X_pca)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = X_pca[:,0], y = X_pca[:,1], hue = customer_df['DBScan Cluster'], palette = 'Set1', legend = 'full')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('DBScan')
plt.show()
```

# Hierarchical Clustering

```python
hierarchical = AgglomerativeClustering(n_clusters = 3)
customer_df['Hierarchical Cluster'] = hierarchical.fit_predict(X_pca)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = X_pca[:,0], y = X_pca[:,1], hue = customer_df['Hierarchical Cluster'], palette = 'Set1', legend = 'full')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Hierarchical')
plt.show()
```

# Gaussian Mixture

```python
gmm = GaussianMixture(n_components = 3, random_state = 42)
customer_df['GMM Cluster'] = gmm.fit_predict(X_pca)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = X_pca[:,0], y = X_pca[:,1], hue = customer_df['GMM Cluster'], palette = 'Set1', legend = 'full')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Gaussian')
plt.show()
```

```python
spectral = SpectralClustering(n_clusters = 3, random_state = 42)
customer_df['Spectral Cluster'] = spectral.fit_predict(X_pca)
```

```python
plt.figure(figsize = (10,6))
sns.scatterplot(x = X_pca[:,0], y = X_pca[:,1], hue = customer_df['Spectral Cluster'], palette = 'Set1', legend = 'full')
plt.xlabel('PCA 1')
plt.ylabel('PCA 2')
plt.title('Gaussian')
plt.show()
```
