```python
import numpy as np
import pandas as pd
import pickle
from datetime import datetime
from sqlalchemy import create_engine, MetaData, Table, Column, Integer, String, Float, Date
from sklearn.cluster import KMeans
from sklearn.preprocessing import PowerTransformer
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
import seaborn as sns
from surprise import Dataset, Reader, SVD
```

```python
MERGED_DATA_PATH = '/home/athel/Desktop/Litmus7/order/MergedData.csv'
NEW_MERGED_DATA_PATH = '/home/athel/Desktop/Litmus7/order/new_MergedData.csv'

ORDER_INFO_NORM = '/home/athel/Desktop/Litmus7/order/order_info_normalized.csv'
ORDER_LINE_NORM = '/home/athel/Desktop/Litmus7/order/order_line_normalized.csv'
PRODUCT_NORM = '/home/athel/Desktop/Litmus7/order/product_normalized.csv'
```

```python
df = pd.read_csv(MERGED_DATA_PATH, parse_dates=['Date'])
```

```python
df['Sales'] = df['Quantity'] * df['Price per Unit']
```

```python
today = pd.to_datetime(datetime.now().date())
df['Recency'] = (today - df['Date']).dt.days
```

```python
df = df.sort_values(['Customer ID', 'Date'])
df['Order Gap'] = df.groupby('Customer ID')['Date'].diff().dt.days
df['Order Gap'] = df['Order Gap'].fillna(0)
```

```python
try:
    new_df = pd.read_csv(NEW_MERGED_DATA_PATH, parse_dates=['Date'])
    # Remove duplicates (by Order ID, Product ID, SKU ID)
    key_cols = ['Order ID', 'Product ID', 'SKU ID']
    merged = new_df.merge(df[key_cols], on=key_cols, how='left', indicator=True)
    new_unique = merged[merged['_merge'] == 'left_only'].drop('_merge', axis=1)
    if not new_unique.empty:
        df = pd.concat([df, new_unique], ignore_index=True)
        print(f"Appended {len(new_unique)} new rows from {NEW_MERGED_DATA_PATH}")
except FileNotFoundError:
    pass
```

```python
# 3.1 order_info: order-level data
order_info_cols = [
    'Order ID', 'Customer ID', 'Warehouse ID', 'Customer Age',
    'Customer Gender', 'Date', 'Recency', 'Order Gap', 'KMeans Cluster'
]
order_info = df[order_info_cols].drop_duplicates().reset_index(drop=True)
order_info.to_csv(ORDER_INFO_NORM, index=False)

# 3.2 order_line: order-product-level data (with price per unit and sales for historical accuracy)
order_line_cols = [
    'Order ID', 'Product ID', 'SKU ID', 'Category',
    'Quantity', 'Price per Unit', 'Sales'
]
order_line = df[order_line_cols].drop_duplicates().reset_index(drop=True)
order_line.to_csv(ORDER_LINE_NORM, index=False)

# 3.3 product: unique products/SKUs
product_cols = ['Product ID', 'SKU ID', 'Category', 'Price per Unit']
product = df[product_cols].drop_duplicates().reset_index(drop=True)
product.to_csv(PRODUCT_NORM, index=False)
```

```python
customer_df = df.groupby('Customer ID').agg(
    total_spend = ('Sales', 'sum'),
    purchase_frequency = ('Order ID', 'nunique'),
    avg_basket_size = ('Quantity', 'mean'),
    cat_diversity = ('Category','nunique'),
    recency = ('Date', lambda x: (pd.Timestamp.today() - x.max()).days),
    gap = ('Order Gap', lambda x: x.sum() / (np.count_nonzero(x) + 1) if np.count_nonzero(x) >= 0 else None),
    age = ('Customer Age', 'first')
).reset_index()

print(df[df['Customer ID'] == 'CUST001']['Order Gap'])
customer_df
```

```python
features = ['total_spend','purchase_frequency','avg_basket_size','cat_diversity','recency','gap','age']
X = customer_df[features].values
X
```

```python
pt = PowerTransformer(method='yeo-johnson', standardize=True)
pca = PCA(n_components=2)
# Fit and transform in sequence
X_pt = pt.fit_transform(X)
X_pca = pca.fit_transform(X_pt)
```

```python
plt.figure(figsize = (15,6))

plt.subplot(1,3,1)
sns.boxplot(x = customer_df['total_spend'])
plt.title('Spend Boxplot')
plt.xlabel('Total Spend')

plt.subplot(1,3,2)
plt.boxplot(x = customer_df['purchase_frequency'])
plt.title('Frequency Boxplot')
plt.xlabel('Frequency')


plt.subplot(1,3,3)
plt.boxplot(x = customer_df['recency'])
plt.title ('Recency Boxplot')
plt.xlabel('Recency')

plt.tight_layout()
plt.show()

customer_df.nsmallest(1, 'recency')
```

# KMeans

```python
kmeans = KMeans(n_clusters=3, random_state=42)
customer_df['KMeans Cluster'] = kmeans.fit_predict(X_pca)

# Plot
plt.figure(figsize=(10,6))
sns.scatterplot(x=X_pca[:,0], y=X_pca[:,1], hue=customer_df['KMeans Cluster'], palette='viridis')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('KMeans')
plt.show()

print(df)
```

```python
with open('pt.pkl','wb') as f:
    pickle.dump(pt, f)
with open('pca.pkl','wb') as f:
    pickle.dump(pca,f)
with open('kmeans_model.pkl','wb') as f:
    pickle.dump(kmeans, f)
```

```python
df = pd.merge(df, customer_df[['Customer ID', 'KMeans Cluster']], on = 'Customer ID', how = 'inner', suffixes = ('_original','_from_customer'))
customer_df.to_csv('customer_df.csv', index = False)
```

```python
for cluster_id in customer_df['KMeans Cluster'].unique():
    cluster_users = customer_df[customer_df['KMeans Cluster'] == cluster_id]['Customer ID']
    cluster_data = df[df['Customer ID'].isin(cluster_users)]
    data_long = cluster_data[['Customer ID','Product ID','Quantity']]
    reader = Reader(rating_scale=(data_long['Quantity'].min(), data_long['Quantity'].max()))
    data = Dataset.load_from_df(data_long, reader)
    trainset = data.build_full_trainset()
    algo = SVD()
    algo.fit(trainset)
    with open(f'cf_model_cluster_{cluster_id}.pkl', 'wb') as f:
        pickle.dump(algo, f)
```
